###### 3) split data


```python
from sklearn.model_selection import train_test_split

X=data.drop("value/offer id",axis=1)
y=data["value/offer id"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=42,stratify=y)
```

###### 4) bulid function contains all steps of preproessing  and polynomail Features( Pipeline)



```python
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline


X_train = X_train.drop(columns=['became_member'], errors='ignore')
X_test  = X_test.drop(columns=['became_member'], errors='ignore')

cat = X_train.select_dtypes(include='object').columns
num = X_train.select_dtypes(exclude='object').columns

encoder = OneHotEncoder(handle_unknown='ignore')
scaler = StandardScaler()

ct = ColumnTransformer([
    ('encoder', encoder, cat),
    ('scaler', scaler, num)
])

X_train_pre = ct.fit_transform(X_train)
X_test_pre  = ct.transform(X_test)
```

```python
df_train=pd.DataFrame(X_train_pre,columns=ct.get_feature_names_out())
```

###### 8) Use Bagging method

```python
####from sklearn.ensemble import BaggingClassifier
#bagg with only KNN model

###parmaeter={"n_estimators":[10,100,500,1000],
     ###          "max_samples":range(31,50,2)}
###knn_bagg=GridSearchCV(estimator=BaggingClassifier(estimator=knn),param_grid=parmaeter,cv=3)

#fit
##knn_bagg.fit(X_train_pre,y_train)


from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, f1_score

# -----------------------------
# Base model: Decision Tree
# -----------------------------
base_tree = DecisionTreeClassifier(
    max_depth=None,
    min_samples_split=5,
    random_state=42
)

# -----------------------------
# Bagging model
# -----------------------------
bag_model = BaggingClassifier(
    estimator=base_tree,
    n_estimators=50,
    max_samples=0.8,
    random_state=42,
    n_jobs=-1
)

# -----------------------------
# Train
# -----------------------------
bag_model.fit(X_train_pre, y_train)

# -----------------------------
# Predict
# -----------------------------
y_pred = bag_model.predict(X_test_pre)
```

```python
#best parameter
####print("Best score in Bagging (Accuracy):", accuracy_score(y_test, y_pred))
#####print("Best score in Bagging (F1 Macro):", f1_score(y_test, y_pred, average='macro'))
# -----------------------------
# Evaluation
# -----------------------------
print("Best parameters in Bagging:")
print({
    "base_estimator": "DecisionTreeClassifier",
    "n_estimators": bag_model.n_estimators,
    "max_samples": bag_model.max_samples,
    "min_samples_split": base_tree.min_samples_split
})

print("\nBest score in Bagging:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Macro:", f1_score(y_test, y_pred, average="macro"))
```

```python
#hyperparameter with only decsion tree model
dtc = DecisionTreeClassifier(
    max_depth=None,
    min_samples_split=5,
    random_state=42
)

bag_model = BaggingClassifier(
    estimator=dtc,
    n_estimators=50,
    max_samples=0.8,  
    random_state=42,
    n_jobs=-1
)


#fit
bag_model.fit(X_train_pre, y_train)
```

```python

# Predict
y_pred = bag_model.predict(X_test_pre)

# Print parameters used
print("Best parameters in Bagging:")
print({
    "base_estimator": "DecisionTreeClassifier",
    "max_depth": dtc.max_depth,
    "min_samples_split": dtc.min_samples_split,
    "n_estimators": bag_model.n_estimators,
    "max_samples": bag_model.max_samples
})

# Print scores
print("\nBest score in Bagging:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Macro:", f1_score(y_test, y_pred, average="macro"))
```
